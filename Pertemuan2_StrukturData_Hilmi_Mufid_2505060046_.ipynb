```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

# create node
n1 = Node("Januari")
n2 = Node("Februari")

class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add newnode in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while(last.nextval):
            last = last.nextval
        last.nextval = NewNode

Li = LinkedList()
Li.headval = n1

# Link first Node to second node
Li.headval.nextval = n2

# add node in the end
Li.AddInEnd("The Last")
Li.listprint()
```

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval =  None

class LinkedList:
  def __init__(self):
    self.headval = None

  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

  # Create atribut add begining to Update the new nodes next val to existing node
  def AddBegining(self,newdata):
    NewNode = Node(newdata)
    NewNode.nextval = self.headval
    self.headval = NewNode

  # Fungsi untuk menghapus node by key
  def RemoveNode(self, Removekey):
    HeadVal = self.headval

    # Case 1: Head node itself holds the key to be deleted
    if HeadVal is not None:
      if HeadVal.dataval == Removekey:
        self.headval = HeadVal.nextval  # Corrected from .next to .nextval
        HeadVal = None
        return

    # Case 2: Search for the key to be deleted, keep track of the previous node
    prev = None
    while HeadVal is not None:
      if HeadVal.dataval == Removekey:
        break # Found the node to delete
      prev = HeadVal
      HeadVal = HeadVal.nextval

    # If key was not present in linked list
    if HeadVal is None:
      return

    # Unlink the node from linked list
    prev.nextval = HeadVal.nextval
    HeadVal = None


# create linked list named Li
Li = LinkedList()

# add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
print("Sebelum dilakukan remove")
Li.listprint()
print("===============================================")

# remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret' ")
Li.listprint()
```

HAPUS DI AKHIR

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add newnode in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while(last.nextval):
            last = last.nextval
        last.nextval = NewNode

    # Function to remove last node
    def RemoveEnd(self):
        last = self.headval
        while (last is not None):
            if last.nextval == None:
                break
            prev = last
            last = last.nextval
        if (last == None):
            return
        prev.nextval = last.nextval
        last = None

# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
print("Sebelum dilakukan remove")
Li.listprint()
print("==========================================")

# remove last node
Li.RemoveEnd()
print("Setelah dilakukan remove node terakhir")
Li.listprint()
```

```python
# NOMER 1
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        """Tambah node di akhir."""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node
        new_node.prev = curr

    def insert_middle(self, target, newdata):
        """Sisipkan node baru setelah node dengan nilai target."""
        curr = self.head
        while curr and curr.data != target:
            curr = curr.next
        if curr is None:
            print(f"Node '{target}' tidak ditemukan.")
            return
        new_node = Node(newdata)
        new_node.next = curr.next
        new_node.prev = curr
        if curr.next:
            curr.next.prev = new_node
        curr.next = new_node

    def delete_by_key(self, key):
        """Hapus node berdasarkan nilai (kata kunci)."""
        curr = self.head
        while curr and curr.data != key:
            curr = curr.next
        if curr is None:
            print(f"Node '{key}' tidak ditemukan.")
            return
        if curr.prev:
            curr.prev.next = curr.next
        else:
            self.head = curr.next   # hapus head
        if curr.next:
            curr.next.prev = curr.prev

    def display(self):
        curr = self.head
        nodes = []
        while curr:
            nodes.append(curr.data)
            curr = curr.next
        print(" ↔ ".join(nodes) + " ↔ None")


# ── DEMO ──────────────────────────────────
dll = DoublyLinkedList()
dll.append("Apple")
dll.append("Banana")
dll.append("Cherry")
print("Awal:"); dll.display()

dll.insert_middle("Apple", "Avocado")
print("Setelah sisip 'Avocado' setelah 'Apple':"); dll.display()

dll.delete_by_key("Banana")
print("Setelah hapus 'Banana':"); dll.display()
```

```python
# NOMER 2
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        """Tambah node di akhir, next node terakhir = head."""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head  # tunjuk ke dirinya sendiri
            return
        curr = self.head
        while curr.next != self.head:
            curr = curr.next
        curr.next = new_node
        new_node.next = self.head    # node baru tunjuk ke head

    def remove_end(self):
        """Hapus node terakhir dari circular linked list."""
        if self.head is None:
            print("List kosong."); return
        if self.head.next == self.head:  # hanya 1 node
            self.head = None; return
        curr = self.head
        while curr.next.next != self.head:
            curr = curr.next
        curr.next = self.head           # node ke-2 dari akhir tunjuk ke head

    def display(self):
        if self.head is None:
            print("List kosong."); return
        nodes = []
        curr = self.head
        while True:
            nodes.append(curr.data)
            curr = curr.next
            if curr == self.head:
                break
        print(" → ".join(nodes) + " → (kembali ke head)")


# ── DEMO ──────────────────────────────────
cll = CircularLinkedList()
cll.append("Senin")
cll.append("Selasa")
cll.append("Rabu")
print("Awal:"); cll.display()

cll.append("Kamis")
print("Setelah sisip 'Kamis' di akhir:"); cll.display()

cll.remove_end()
print("Setelah hapus node terakhir:"); cll.display()
```
